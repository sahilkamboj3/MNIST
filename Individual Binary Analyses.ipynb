```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
import numpy as np

from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.linear_model import SGDClassifier
from sklearn.base import clone
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, precision_recall_curve, roc_curve, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

mnist = datasets.load_digits()
```

```python
X = mnist['data']
y = mnist['target']
```

```python
classes = set(y)
```

# Data Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
shuffle_index = np.random.permutation(len(X_train))
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
```

# Helper Functions

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[: -1], 'b--', label='Precision')
    plt.plot(thresholds, recalls[: -1], 'g--', label='Recall')
    plt.xlabel('Threshold')
    plt.legend(loc='lower left')
    plt.ylim([-0.5,1.25])
```

```python
def plot_roc_curve(fpr, tpr, label=None): # good classifier tries to stay as far in the top left as possible
    plt.plot(fpr, tpr, linewidth=1, label=label)
    plt.plot([0,1], [0,1], 'k--')
    plt.axis([0,1,0,1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
```

# Binary Models

```python
fig = plt.figure(1, figsize=(30, 320))
i = 0

for number in classes:
    print(number)
    cur_y_train = (y_train == number)
    
    print('SGD')
    print('---')
    sgd = SGDClassifier(random_state=42)
    
    # Cross_Val_Score
    sgd_scores = cross_val_score(sgd, X_train, cur_y_train, cv=5, scoring='accuracy')
    print(f'Cross_Val_Scores: {sgd_scores}')
    print(f'Cross_Val_Scores Mean: {sgd_scores.mean()}')
    
    # Cross_Val_Predict
    sgd_pred = cross_val_predict(sgd, X_train, cur_y_train, cv=5)
    sgd_conf_matrix = confusion_matrix(sgd_pred, cur_y_train)
    print(f'SGD_Matrix: {sgd_conf_matrix}')
    
    pres_score = precision_score(cur_y_train, sgd_pred) 
    rec_score = recall_score(cur_y_train, sgd_pred)
    print(f'Precision score: {pres_score}')
    print(f'Recall score: {rec_score}')
    
    # Recall/Precision vs Threshold
    i += 1
    ax1 = plt.subplot(30, 3, i)
    
    precisions, recalls, thresholds = precision_recall_curve(cur_y_train, sgd_pred)

    ax1.plot(thresholds, precisions[: -1], 'b--', label='Precision')
    ax1.plot(thresholds, recalls[: -1], 'g--', label='Recall')
    ax1.set_xlabel('Threshold', fontsize=15)
    ax1.set_title(f'{number} - SGD Precision/Recall over Threshold', fontsize=15)
    ax1.legend(loc='lower left')
    ax1.set_ylim([-0.5,1.25])
    plt.xticks(fontsize=15)
    
    # Precision vs Recall
    i += 1
    ax2 = plt.subplot(30, 3, i)
    
    ax2.plot(precisions[:-1], recalls[: -1])
    ax2.set_xlabel('Precision', fontsize=15)
    ax2.set_ylabel('Recall', fontsize=15)
    ax2.set_title(f'{number} - SGD Precision vs Recall', fontsize=15)
    plt.xticks(fontsize=15)
    
    # ROC Curve
    fpr, tpr, thresholds = roc_curve(cur_y_train, sgd_pred)
    print(f'ROC_AUC_Score: {roc_auc_score(cur_y_train, sgd_pred)}') # Area under the curve.
    
    #### Random Forest Classifier
    print('')
    print('Random Forest')
    print('-------------')
    
    rfc = RandomForestClassifier(random_state=42)
    
    # Cross_Val_Scores
    rfc_scores = cross_val_score(rfc, X_train, cur_y_train, cv=5, scoring='accuracy')
    print(f'Cross_Val_Scores: {rfc_scores}')
    print(f'Cross_Val_Scores Mean: {rfc_scores.mean()}')
    
    # Cross_Val_Predict
    rfc_pred = cross_val_predict(rfc, X_train, cur_y_train, cv=5)

    rfc_conf_matrix = confusion_matrix(cur_y_train, rfc_pred)
    print(f'RFC_Matrix: {rfc_conf_matrix}')
    
    pres_score = precision_score(cur_y_train, rfc_pred) 
    rec_score = recall_score(cur_y_train, rfc_pred)
    print(f'Precision score: {pres_score}')
    print(f'Recall score: {rec_score}')
    
    rfc_pred_2 = cross_val_predict(rfc, X_train, cur_y_train, cv=5, method='predict_proba')
    y_scores_forest = rfc_pred_2[:, 1]
    
    fpr_forest, tpr_forest, thresholds = roc_curve(cur_y_train, y_scores_forest)
    
    i += 1
    ax3 = plt.subplot(30, 3, i)    
    ax3.plot(fpr, tpr, 'b:', label='SGD')
    ax3.plot(fpr_forest, tpr_forest, linewidth=1, label='Random Forest')
    ax3.plot([0,1], [0,1], 'k--')
    ax3.set_xlabel('False Positive Rate', fontsize=15)
    ax3.set_ylabel('True Positive Rate', fontsize=15)
    ax3.set_title(f'{number} - RFC TPR vs FPR', fontsize=15)
    plt.xticks(fontsize=15)
    
    print(roc_auc_score(cur_y_train, y_scores_forest))
    print('')
```
